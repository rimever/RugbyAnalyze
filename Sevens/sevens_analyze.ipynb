```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

data = pd.read_csv( 'cleaned_data.csv',comment ='#')
data.head()
```

```python
new_data = pd.DataFrame()



for index, row in data.iterrows():        
    values = []
    
    if row['LeftPoint'] > row['RightPoint']:
        values.append(row['LeftTeam'])
        values.append(row['RightTeam'])
        values.append(row['LeftPoint'])
        values.append(row['RightPoint'])
    else:
        values.append(row['RightTeam'])
        values.append(row['LeftTeam'])
        values.append(row['RightPoint'])
        values.append(row['LeftPoint'])

    print(values)
        
    series = pd.Series(values, index=data.columns)
    new_data = new_data.append(series, ignore_index = True)

new_data.describe()
```

```python
points = (new_data['LeftPoint'], new_data['RightPoint'], new_data['LeftPoint'] - new_data['RightPoint'])

rugby_union_data = pd.read_csv( '../Match/match_data_cleaning.csv',comment ='#')




# 箱ひげ図
fig,( axL,axR) = plt.subplots(ncols=2,figsize=(10,8),sharex=True)
axL.boxplot(points)
axL.set_xticklabels(['勝利チーム', '敗北チーム','点差'])

axL.set_title('セブンズ（７人制）')
axL.set_xlabel('team')
axL.set_ylabel('point')
axL.set_ylim(0,150)


rugby_union_point = (rugby_union_data['For'],rugby_union_data['Aga'],rugby_union_data['For']-rugby_union_data['Aga'])
axR.boxplot(rugby_union_point)
axR.set_xticklabels(['勝利チーム', '敗北チーム','点差'])

axR.set_title('ラグビーユニオン（１５人制）')
axR.set_xlabel('team')
axR.set_ylabel('point')
axR.set_ylim(0,150)


# 描画
plt.show()
```
