```python
## pandasを使って収集したデータについて軽く調べてみます。
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import re
import os


data = pd.read_csv('player_world_stats.csv', index_col=0)
data = data.drop_duplicates(subset='name')
temp_born = []
for born in data.Born:
    result = re.search('1\d{3}', str(born))
    if result == None:
        temp_born.append(0)
    else:
        temp_born.append(int(result.group()))

data["Born"] = temp_born

data.head()
```

```python
data['Rate'] =  data['Pts'] / data['Mat']
points = data['Rate']
matches = data['Mat']

plt.xlabel('Mat')
plt.ylabel('Rate')

plt.scatter(matches,points)
```

```python
data['Rate'] =  data['Pts'] / data['Mat']
result = data.query('Rate > 10 & Mat > 40')
temp = result[['name','Rate','Pts','Mat','Span','Born']]
temp = temp.sort_values('Rate', ascending=False)
temp
```
