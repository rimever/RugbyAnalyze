```python
import urllib.request
import bs4
# ESPNのニュージーランドの選手一覧ページにアクセスします。
#url = 'http://en.espn.co.uk/newzealand/rugby/player/caps.html?team=8'
#soup = bs4.BeautifulSoup(urllib.request.urlopen(url).read())
#print(str(soup))

path = "players_link.txt"

with open(path) as f:
    l = f.readlines()

link = l[0].replace('\n','')

url = 'http://en.espn.co.uk/' + link
soup = bs4.BeautifulSoup(urllib.request.urlopen(url).read())
print(str(soup))
```

```python
data = soup.find("div", class_="scrumPlayerName")

stats = {}
stats["name"] = data.string

print(stats)
```

```python
data = soup.find_all("div", class_="scrumPlayerDesc")
for item in data:
    try:
        key = item.find("b").string
        text = str(item)
        start_index = text.find("</b>")
        end_index = text.find("</div>")
        value = text[start_index + 4:end_index].strip()
        stats[key] = value
    except AttributeError:
        pass
    
print(stats)
```

```python
data = soup.find("thead").find_all("th")
head =[]
for item in data:
    head.append(item.string)

print(head)
```

```python
data = soup.find("table", class_="engineTable").find_all("tr")
for item in data:
    if item.find("b").string == "All Tests":
        for index, table_data in enumerate(item.find_all("td")):
            if index > 0:
                stats[head[index]] =  table_data.string

print(stats)
```

```python
import pandas as pd
data_frame = pd.DataFrame(columns = stats.keys())
values = []
for v in stats.values():
    values.append(v)
series = pd.Series(values, index=data_frame.columns)
data_frame.append(series, ignore_index = True)
```
