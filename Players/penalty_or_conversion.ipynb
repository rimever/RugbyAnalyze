```python
players = pd.read_csv( 'player_world_stats.csv' ,index_col=0)
players = players.drop_duplicates(subset='name')

players = players[['name', 'Born','Span','Position','Mat','Start','Sub','Pts','Tries','Conv','Pens','Drop','%','Height','Weight']]
```

```python
players['TriesPerMat'] = players['Tries'] / players['Mat']
players['ConvPerMat'] = players['Conv'] / players['Mat']
players['PensPerMat'] = players['Pens'] / players['Mat']
players['ConvByPens'] = players['Conv'] / players['Pens']
```

```python
xlabel_name = 'Conv'
ylabel_name = 'Pens'

x_data = players[xlabel_name]
y_data = players[ylabel_name]
plt.xlabel('コンバージョン')
plt.ylabel('ペナルティゴール')
# コンバージョンとペナルティーゴールを同列に扱いたいので範囲は揃える
plt.xlim(0,300)
plt.ylim(0,300)

plt.scatter(x_data,y_data)
```

```python
temp = players.query("Mat > 40")
xlabel_name = 'ConvPerMat'
ylabel_name = 'PensPerMat'

X = temp[[xlabel_name]]
Y = temp[ylabel_name]

#サンプルをランダムに分けるcross_validation
#import sklearn.cross_validation 
import sklearn.model_selection
from sklearn.linear_model import LinearRegression

#左辺にモデル構築用のXY,テスト用のX,Yを用意します。
#先ほど作成した、説明変数を第一引数に、目的変数を第二引数に置きます。
#train_test_splitを足すことを忘れないでください！！
X_train, X_test, Y_train, Y_test = sklearn.model_selection.train_test_split(X,Y)

#線形回帰モデルを作成
multi_lreg = LinearRegression()

#X,Y双方がtrain用のデータである点にご注意ください。
multi_lreg.fit(X_train, Y_train)

#トレーニングに使用したデータセットでの精度
print(multi_lreg.score(X_train,Y_train))

#テストに使用したデータセットでの精度
print(multi_lreg.score(X_test,Y_test))
```

```python
print(multi_lreg.coef_)
```
